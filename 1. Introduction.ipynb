## MLOps
ML을 개발하고 서비스화와 유연하게 운영하는 데 필요한 모든 것을 다루는 분야이다.
- **데이터**를 기반으로 프로그램을 개발함으로 ML과 DevOps 기술을 결합해 사용한다.

## MLOps 필요성
- 학습/개발환경에서의 성능을 배포환경에서 재현하지 못하는 경우가 다반사로 성능이 좋았던 순서대로 다른 모델을 올려보아야 한다.
- 협업자간 개발환경이 다르며 데이커 크기, GPU, 서버 사용 방식에 따라 타인 혹은 과거의 성능 재현이 어렵다.
- Gartner 보고에 따르면 85%의 AI/ML 프로그램이 이러한 문제로 서비스화 되지 못하고 있다고 한다.<br>
  [출처] https://www.infoworld.com/article/3639028/why-ai-investments-fail-to-deliver.html
  
  
- 서비스화에서 ML모델 개발에만 치중하면 안되고 성능 저하(AB test), 시스템 오류 발생, 데이터 분포 변화 모니터링 및 Outlier 검증, AutoML 수행 등을 자동적으로 수행되도록 설계해야 한다. <br>
  
  
  [출처] https://cloud.google.com/architecture/mlops-continuous-delivery-and-automation-pipelines-in-machine-learning?hl=ko
